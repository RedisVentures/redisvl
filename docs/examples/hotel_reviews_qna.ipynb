```python
import pandas as pd
df = pd.read_csv('https://query.data.world/s/j3adtsuf5mjtknqxvpfoj4u6csl64k?dws=00000')
df.head()
```

```python
cols = ['name', 'latitude', 'longitude', 'address', 'city', 'province', 'reviews.text', 'reviews.title', 'reviews.rating', 'country', 'categories']

clean_data = df[cols].copy()
#clean_data["location"] = clean_data["latitude"].astype(str) + "," + clean_data["longitude"].astype(str)
clean_data.rename(
    columns={
        "reviews.text": "review",
        "reviews.rating": "rating",
        "reviews.title": "title",
        "reviews.date": "date",
        "province": "state"
    },
    inplace=True)

clean_data.drop(columns=["latitude", "longitude"], inplace=True)
clean_data.title.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
clean_data.review.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
clean_data.dropna(inplace=True)
clean_data.title.apply(lambda x: clean_description(x))
clean_data.title.fillna("no-title", inplace=True)
clean_data["rating"] = clean_data["rating"].astype(int)
clean_data.head()
```

```python
import re
import string

# Helper function to clean the description!
def clean_description(description: str):
    if not description:
        return ""
    # remove unicode characters
    description = description.encode('ascii', 'ignore').decode()

    # remove punctuation
    description = re.sub('[%s]' % re.escape(string.punctuation), ' ', description)

    # clean up the spacing
    description = re.sub('\s{2,}', " ", description)

    # remove urls
    #description = re.sub("https*\S+", " ", description)

    # remove newlines
    description = description.replace("\n", " ")

    # remove all numbers
    #description = re.sub('\w*\d+\w*', '', description)

    # split on capitalized words
    description = " ".join(re.split('(?=[A-Z])', description))

    # clean up the spacing again
    description = re.sub('\s{2,}', " ", description)

    # make all words lowercase
    description = description.lower()

    return description.strip()
```

# QNA System

- optionally pick a city
- pick a radius from within to search from
- use hyde approach to generate a fake review
- use the fake review to query the QNA system to look for reviews
- write a prompt to include reviews and generate recommendations based on the review

flow
1. What city are you looking to stay in?
2. How far from the city center do you want to stay?
3. what qualities are you looking for in a hotel?
4. what qualities are you looking to avoid in a hotel?



```python
import openai
from redisvl.query import VectorQuery
```

```python
import getpass

openai.api_key = getpass.getpass()
```

```python
CHAT_MODEL = "gpt-3.5-turbo"

def make_hyde_review(positive, negative):
    retrieval_prompt = f'''Your job is to generate a review for a hotel based on the positive and negative qualities provided.
    The review should be at least 10 words long and from the prospective of a customer who stayed at the hotel. Be informal
    and consise. You aren't that smart.

    Positive Qualities the user would like:

    {positive}

    Negative Qualities the user would like to avoid:

    {negative}
    '''
    return retrieval_prompt


def make_prompt(positive, negative, reviews):
    retrieval_prompt = f'''You are a service dedicated to recommending hotels based on user reviews.
    You will be provided positive and negative qualities the user is looking for in a hotel, as well as a
    large number of reviews. You will then be asked to recommend a hotel based on the user's preferences and explain why.
    Always start the suggestion with "Based on user reviews, I suggest the following hotel:"

    Positive Qualities the user would like:

    {positive}

    Negative Qualities the user would like to avoid:

    {negative}

    Reviews:

    {reviews}

    Format for your response:

    Hotel: <hotel name>
    Reason for Suggestion: <reason for suggestion>
    Other options:
        1. <hotel name>
        2. <hotel name>
        3. <hotel name>
    '''
    return retrieval_prompt

def retrieve_context(positive, negative):

    hyde_prompt = make_hyde_review(positive, negative)

    hyde_review = openai.ChatCompletion.create(
        model=CHAT_MODEL,
        messages=[{'role':"user",
                   'content': hyde_prompt}],
        max_tokens=200)

    hyde_review = hyde_review['choices'][0]['message']['content']

      # Embed the query
    query_embedding = hf.embed(hyde_review)

    # Get the top result from the index
    vector_query = VectorQuery(
        vector=query_embedding,
        vector_field_name="embedding",
        return_fields=["review", "name", "title"],
        num_results=30
    )

    results = index.query(vector_query)
    content = []
    if len(results.docs) > 1:
        # join the hotel name and the review
        for doc in results.docs:
            content.append(f"Hotel Name: {doc['name']}\n Review Title: {doc['title']}\n Review: {doc['review']}\n")
    return "\n".join(content)


def find_hotel(postive, negative):

    # Retrieve the context
    reviews = retrieve_context(positive, negative)

    prompt = make_prompt(positive, negative, reviews)
    retrieval = openai.ChatCompletion.create(
        model=CHAT_MODEL,
        messages=[{'role':"user",
                   'content': prompt}],
        max_tokens=1000)

    # Response provided by GPT-3.5
    return retrieval['choices'][0]['message']['content'], reviews
```
